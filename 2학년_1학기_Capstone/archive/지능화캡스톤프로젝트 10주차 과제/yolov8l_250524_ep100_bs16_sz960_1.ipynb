```python
# ì••ì¶•í•  í´ë”ì˜ ê²½ë¡œë¥¼ ë³€ìˆ˜ì— ì €ì¥í•©ë‹ˆë‹¤.
# ì—¬ê¸°ì— ì••ì¶•í•˜ê³  ì‹¶ì€ í´ë”ì˜ ì •í™•í•œ ê²½ë¡œë¥¼ ì…ë ¥í•´ ì£¼ì„¸ìš”.
# ì˜ˆì‹œ: '/content/drive/MyDrive/capstone/yolov8'
source_folder_path = '/content/runs'

# ì••ì¶• íŒŒì¼ì´ ì €ì¥ë  ê²½ë¡œì™€ íŒŒì¼ ì´ë¦„ì„ ì§€ì •í•©ë‹ˆë‹¤.
# ì—¬ê¸°ì„œëŠ” ì½”ë©ì˜ ì„ì‹œ ì €ì¥ ê³µê°„ì— 'yolov8_archive.zip'ì´ë¼ëŠ” ì´ë¦„ìœ¼ë¡œ ì €ì¥ë˜ë„ë¡ ì„¤ì •í•©ë‹ˆë‹¤.
# ë‚˜ì¤‘ì— ì´ íŒŒì¼ì„ ë¡œì»¬ë¡œ ë‹¤ìš´ë¡œë“œí•  ê²ƒì…ë‹ˆë‹¤.
output_zip_path = '/content/yolov8l_ep100_bs16_sz960.zip'

# zipfile ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ì‚¬ìš©í•˜ì—¬ í´ë”ë¥¼ ì••ì¶•í•©ë‹ˆë‹¤.
# zipfileì€ íŒŒì´ì¬ì— ê¸°ë³¸ìœ¼ë¡œ ë‚´ì¥ë˜ì–´ ìˆì–´ì„œ ë”°ë¡œ ì„¤ì¹˜í•  í•„ìš”ê°€ ì—†ìŠµë‹ˆë‹¤.
import zipfile
import os

print(f"'{source_folder_path}' í´ë” ì••ì¶•ì„ ì‹œì‘í•©ë‹ˆë‹¤...")

try:
    # 'w' ëª¨ë“œë¡œ íŒŒì¼ì„ ì—´ë©´ ìƒˆë¡œìš´ ZIP íŒŒì¼ì„ ìƒì„±í•©ë‹ˆë‹¤.
    with zipfile.ZipFile(output_zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # os.walkë¥¼ ì‚¬ìš©í•˜ì—¬ í´ë”ì™€ í•˜ìœ„ í´ë”, íŒŒì¼ë“¤ì„ ëª¨ë‘ íƒìƒ‰í•©ë‹ˆë‹¤.
        for root, dirs, files in os.walk(source_folder_path):
            for file in files:
                # ê° íŒŒì¼ì˜ ì „ì²´ ê²½ë¡œë¥¼ ë§Œë“­ë‹ˆë‹¤.
                file_path = os.path.join(root, file)
                # ZIP íŒŒì¼ ë‚´ë¶€ì— ì €ì¥ë  ê²½ë¡œë¥¼ ì§€ì •í•©ë‹ˆë‹¤.
                # source_folder_pathì˜ ê¸¸ì´ë¥¼ ë¹¼ì„œ ìƒëŒ€ ê²½ë¡œë¡œ ë§Œë“­ë‹ˆë‹¤.
                # ì´ë ‡ê²Œ í•´ì•¼ ì••ì¶• í•´ì œ ì‹œ 'capstone/yolov8/...' ì²˜ëŸ¼ í’€ë¦½ë‹ˆë‹¤.
                arcname = os.path.relpath(file_path, source_folder_path)
                # íŒŒì¼ì„ ZIP íŒŒì¼ì— ì¶”ê°€í•©ë‹ˆë‹¤.
                zipf.write(file_path, arcname)
    print(f"'{source_folder_path}' í´ë”ê°€ '{output_zip_path}' (ìœ¼)ë¡œ ì„±ê³µì ìœ¼ë¡œ ì••ì¶•ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print('ì´ì œ ì••ì¶•ëœ íŒŒì¼ì„ ë¡œì»¬ë¡œ ë‹¤ìš´ë¡œë“œí•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.')

except Exception as e:
    print(f"ì••ì¶• ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")
```
