```python
from ultralytics import YOLO
import cv2
import os

# ✅ 모델 로딩
model_path = "./yolo12x.pt"
model = YOLO(model_path)

# ✅ 비디오 및 저장 폴더
video_path = "/content/drive/MyDrive/capstone/정푸드코리아_ch11_20250512114500_20250512131429.mp4의 사본"
save_dir = "./정푸드코리아_ch11_20250512114500_20250512131429"
os.makedirs(save_dir, exist_ok=True)

# ✅ 비디오 읽기
cap = cv2.VideoCapture(video_path)
frame_idx = 0
saved_count = 0

# ✅ 최대 10장만 추출
while cap.isOpened() and saved_count < 10:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_idx % 30 == 0:
        results = model.predict(source=frame, classes=[0], verbose=False)

        if results and results[0].boxes is not None and len(results[0].boxes) > 0:
            save_path = os.path.join(save_dir, f"frame_{frame_idx:06}.jpg")
            cv2.imwrite(save_path, frame)
            saved_count += 1
            print(f"[INFO] Saved: {save_path}")

    frame_idx += 1

cap.release()
print(f"✅ 저장 완료: 총 {saved_count}개의 프레임이 저장되었습니다.")
```

```python
from ultralytics import YOLO
import cv2
import os

# ✅ 모델 및 경로 설정
model_path = "/content/drive/MyDrive/capstone/yolov8_2nd_version_train/yolov8n_ep100_bs16_img640/weights/best.pt"
video_path = "/content/drive/MyDrive/capstone/vid/정푸드코리아_ch11_20250522125957_20250522133008.mp4"

# ✅ 모델 경로 기반 폴더 구성
from pathlib import Path
model_base = Path(model_path).parents[1]  # yolov8_2nd_version_train/yolov8n_ep100_bs16_img960
save_dir = os.path.join("./visual_output_6frames", model_base.parts[-2], model_base.parts[-1])
os.makedirs(save_dir, exist_ok=True)

# ✅ 모델 로딩
model = YOLO(model_path)

# ✅ 클래스별 색상 정의 (최대 10종까지 임의 색상 지정)
colors = {
    0: (0, 255, 0),     # person - green
    1: (255, 0, 0),     # gown_on - blue
    2: (0, 0, 255),     # hairnet_on - red
    3: (255, 255, 0),   # 기타
    4: (255, 0, 255)
}

# ✅ 비디오 읽기
cap = cv2.VideoCapture(video_path)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
selected_frames = 6
interval = total_frames // selected_frames

# ✅ 프레임 추출 루프
saved = 0
frame_idx = 0

while cap.isOpened() and saved < selected_frames:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_idx % interval == 0:
        result = model.predict(source=frame, verbose=False)
        for box in result[0].boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            label_name = model.names[cls]
            label = f"{label_name} {conf:.2f}"
            color = colors.get(cls, (200, 200, 200))  # fallback color: gray

            # ✅ 바운딩 박스 및 라벨
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        # ✅ 이미지 저장
        save_path = os.path.join(save_dir, f"frame_{frame_idx:06}.jpg")
        cv2.imwrite(save_path, frame)
        print(f"✅ 저장됨: {save_path}")
        saved += 1

    frame_idx += 1

cap.release()
```
