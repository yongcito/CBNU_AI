```python
import os

# 이미지와 라벨이 있는 경로 설정
img_dir = "/content/Dataset_Test_Capstone-3/test/images"
label_dir = "/content/Dataset_Test_Capstone-3/test/labels"

# .jpg 확장자를 가진 이미지 파일만 불러오기 (정렬 포함)
image_files = sorted([f for f in os.listdir(img_dir) if f.endswith(".jpg")])

# 결과 확인
print(f"🔍 총 이미지 수: {len(image_files)}장")
print("🖼️ 예시 이미지 파일명:", image_files[:3])
```

```python
from PIL import Image
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for i, ax in enumerate(axes):
    img = Image.open(os.path.join(img_dir, image_files[i]))
    ax.imshow(img)
    ax.set_title(f"{image_files[i]}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
for img_name in image_files[:3]:
    label_name = os.path.splitext(img_name)[0] + '.txt'
    label_path = os.path.join(label_dir, label_name)
    print(f"\n[{label_name}]")
    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            print(f.read().strip())
    else:
        print("라벨 파일 없음")
```
