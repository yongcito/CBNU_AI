```python
import pandas as pd
import numpy as np
from datetime import datetime

# 2020년부터 2023년까지의 날짜 생성
start_date = datetime(2020, 1, 1)
end_date = datetime(2023, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date)

# 월별 출고수량 데이터 생성
monthly_data = []
for year in range(2020, 2024):
    for month in range(1, 13):
        # 코로나 영향으로 20년 출고수량을 낮게 설정
        if year == 2020:
            if month in [6, 7, 8]:
                quantity = np.random.randint(50, 100)
            else:
                quantity = np.random.randint(30, 80)
        # 그 이후의 연도에는 출고수량이 점진적으로 증가
        else:
            if month in [6, 7, 8]:
                quantity = np.random.randint(400, 600)
            elif month in [9, 10, 11]:
                quantity = np.random.randint(300, 450)
            elif month in [12, 1, 2]:
                quantity = np.random.randint(200, 400)
            else:
                quantity = np.random.randint(250, 400)
        monthly_data.append({'연도': year, '월': month, '출고수량': quantity})

# 데이터프레임 생성
df_whelk = pd.DataFrame(monthly_data)

# 데이터셋 출력
print(df_whelk)
```
